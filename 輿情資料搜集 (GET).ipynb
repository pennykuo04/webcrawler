### 搜集新聞網站內即時新聞的標題及內容
url:https://money.udn.com/money/index

```python
# -*- coding: utf-8 -*-
import requests
from bs4 import BeautifulSoup

new_list = []
url = 'https://money.udn.com/money/index'

## headers
headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:108.0) Gecko/20100101 Firefox/108.0'}
res = requests.get(url, headers = headers)
soup = BeautifulSoup(res.text,'html')
# print(soup.prettify())

## 抓取即時區域所有新聞title以及新聞內容連結link，並將每篇新聞的連結link append到new_list內
story = soup.select('div[class="story__list story__list--web"]')[:4]
#print(story)


for i in story:
    try:
        print('title:',i.a['title'])
        print('link:','https://money.udn.com'+i.a['href'])
        new_list.append('https://money.udn.com'+i.a['href'])
    except AttributeError as e:
        print(i)
        print(e.args)
        print('==========')


#print(new_list)
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlretrieve
from time import sleep
import os

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

        
def save_news(url):
    
    headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:108.0) Gecko/20100101 Firefox/108.0'}
    res = requests.get(url, headers = headers)
    sleep(1)
    
    soup = BeautifulSoup(res.text, 'html')
    
    ## 使用find_all去抓取每篇新聞的title，將結果所產生之list放到new_title變數內
    new_title = soup.find_all('h1',{'id':'story_art_title'})
    
    save_path = './news/'+new_title[0].text.replace(':','')
    
    article_body = soup.find_all('section',{'id':'article_body'})
    print('new title: ', new_title[0].text)

    ## create folder
    createFolder(save_path)
    soup2 = BeautifulSoup(str(article_body[0]), 'html')
    
    
    ## save article contens in folder
    with open(save_path+'/contents.txt', 'w', encoding='utf-8') as f:  
        ## 抓取文章內p tags，並將新聞內文寫入contents.txt檔案內
        p_tags = soup2.find_all('p')
        print('new content: ')
        
        for i in p_tags:
            print(i.text)
            f.write(i.text)

    ## save image in article        
    img_tags = soup2.find_all('img')
    
    ## 抓取文章內所有img tags，並將圖片下載並存入對應新聞資料夾下
    print('save image: ')
    for i in img_tags:
        print(i['src'])
        urlretrieve(i['src'], save_path+'/'+i['title']+'.jpg')
    print('==========\n')
    

for new_url in new_list:
    print(new_url)
    save_news(new_url)
    
    
```
